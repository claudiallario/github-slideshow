```
#Ejercicio 1: Libras a gramos
def libras_a_gramos(libras):
    gramos = 453.59237 * libras
    print(libras, "libras son", gramos, "gramos")
    return gramos

#Ejercicio 2: Declinación solar
import math
def declinacion(dia):
    d = 23.45 * math.sin(((2*math.pi)*(dia + 284))/365)
    print(dia, "dias son", d, "º")
    return d

#Ejercicio 3: Área del jardín
import math
def area_jardin(r):
    area_circulo = math.pi * r**2
    area_cuadrado = (r**2)/2
    a = area_circulo - area_cuadrado
    return a

#Ejercicio 4: Momento de inercia de un tubo
def momento_inercia_tubo(m,a,b):
    i = (1/2) * m *(a**2 + b**2)
    return i

#Ejercicio 5: Cardioide
import math
def cardioide(a,theta):
    c = a * (1 + math.cos(theta))
    return c

#Ejercicio 6: Empuje del motor de un cohete
def empuje_cohete(m,veact,ae,pe,pamb):
    e = m * veact + ae*(pe - pamb)
    return float(e)

#Ejercicio 7: Ángulos de un triángulo dados los lados
import math
def angulos_triangulo(a,b,c):
    A = math.acos((b**2 + c**2 - a**2)/(2*b*c))
    B = math.acos((a**2 + c**2 - b**2)/(2*a*c))
    C = math.acos((a**2 + b**2 - c**2)/(2*a*b))
    return A, B, C

#Ejercicio 8: Velocidad de escape
import math
def velocidad_escape(g,r):
    ve = math.sqrt(2*g*r)
    return float(ve)

#Ejercicio 9: Velocidad orbital
import math
G = 6.674 * (10**-11)
def velocidad_orbital(m,r,a):
    vo = math.sqrt(2*G*m*((1/r)-(1/2*a)))
    return float(vo)

#Ejercicio 10: Catenaria
import math
def catenaria(x,a):
  y = a * math.cosh(x/a)
  return float(y)
```
